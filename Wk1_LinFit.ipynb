```python
# This script calculates and plots the linear best fit line to the data found in file "Week1_1_LinFit.csv" using the
# polyfit function and no associated errors

import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('Week1_1_LinFit.csv', delimiter=',')
x = data[:, 0]
y = data[:, 1]

coefficients, cov_matrix = np.polyfit(x, y, deg=1, cov=True)
m, c = coefficients
sigma_m, sigma_c = np.diag(cov_matrix)

# Create the fitted line
fitted_line = m * x + c

# Make plot with error bars
plt.scatter(x, y, label='Data')

# Add fitted line to plot
plt.plot(x, fitted_line, label=f'Fitted Line (m={m:.3f}, c={c:.3f})', color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Show plot
plt.grid(True)
plt.show()

# Print variable values
print("Using the equation y=mx+c:")
print("m = ", m)
print("c = ", c)
print("m_e = ", sigma_m)
print("c_e = ", sigma_c)
```

```python
data = np.loadtxt('Week1_1_LinFit.csv', delimiter=',')
x = data[:, 0]
y = data[:, 1]
sigma_y = data[:, 2]
weights = 1 / sigma_y  # Calculating weights

coefficients, cov_matrix = np.polyfit(x, y, deg=1, w=weights, cov=True)
m, c = coefficients
sigma_m, sigma_c = np.diag(cov_matrix)

# Create the fitted line
fitted_line = m * x + c

errors = (y - fitted_line)/sigma_y
chi_squared = np.sum(errors**2)

# Make plot with error bars
plt.errorbar(x, y, yerr=sigma_y, fmt='o', label='Data with Error Bars')

# Add fitted line to plot
plt.plot(x, fitted_line, label=f'Fitted Line (m={m:.3f}, c={c:.3f})', color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Show plot
plt.grid(True)
plt.show()

# Print variable values
print("Using the equation y=mx+c:")
print("m = ", m)
print("c = ", c)
print("m_e = ", sigma_m)
print("c_e = ", sigma_c)
print("chi^2 = ", chi_squared)
```
